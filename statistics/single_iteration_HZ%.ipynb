```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
plt.rcParams['figure.figsize'] = (15,15)
```

```python
N = 3

a = (700 , 562 , 462)
b = (345000 , 150000 , 150000)
c = (135000 , 44916 , 44916)
d = (466 , 416 , 366)
e = (0 ,115184 , 0)
f = (120000 , 29952 , 60000)
g = (466 , 297 , 373)
h = (230000 , 82009 , 150061)

r = [0,1,2]
Width = 0.35
ind = np.arange(N)

p1 = plt.bar(r , a ,color='#fc0303', edgecolor='white', width=Width)
p2 = plt.bar(r, b, bottom=a, color='#fc8403', edgecolor='white', width=Width)
p3 = plt.bar(r, c, bottom=b, color='#fce703', edgecolor='white', width=Width)
p4 = plt.bar(r, d, bottom=c, color='#b5fc03', edgecolor='white', width=Width)
p5 = plt.bar(r, e, bottom=d, color='#4263f5', edgecolor='white', width=Width)
p6 = plt.bar(r, f, bottom=e, color='#03fc90', edgecolor='white', width=Width)
p7 = plt.bar(r, g, bottom=f, color='#03fcf4', edgecolor='white', width=Width)
p8 = plt.bar(r, h, bottom=g, color='#f542f2', edgecolor='white', width=Width)

plt.ylabel('Power in Watts')
plt.xlabel('Hotzone Percentage')
plt.title('Energy distribution for various hot-zone percentages')
plt.xticks(ind, ('0%','50%','100'))
plt.legend((p1[0], p2[0] , p3[0] , p4[0] , p5[0] , p6[0] , p7[0] , p8[0]), ('Bringing up nodes', 'Initial block-distribution' , 'Initial block access' , 'Maintaining the infrastructure' ,' Transfer data to cold zone' , 'Block balancer' ,' Maintaing the infrastructure ', ' Continued block access'))


plt.show()
```
